```python
class customer:
    idn = ""
    theme = 0
    alpha = 0
    beta = 0
    pilot = 0
    r = 0
    inn = 0
  
    def __init__(self, idn, theme, alpha, beta, pilot, r, inn):
       self.idn = idn
       self.theme = theme
       self.alpha = alpha
       self.beta = beta
       self.pilot = pilot
       self.r = r
       self.inn = inn
    def __repr__(self):
        return "<idn:%s theme:%s alpha:%s beta:%s pilot:%s r:%s inn:%s>" % (self.idn, self.theme, self.alpha, self.beta, self.pilot, self.r, self.inn)

   
```

```python
class startup:
    best_fit = []
    idn = ""
    theme= ""
    def __init__(self, idn, theme):
        self.idn = idn
        self.theme = theme
        self.best_fit = best_fit
```
